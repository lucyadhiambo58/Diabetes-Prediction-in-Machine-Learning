```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

Loading Datasets

```python
x, y = datasets.load_diabetes(return_X_y=True)
```

Print the shape of the data and the first row

```python
print (x.shape)
print (x[0])
```

Extract the column at index 2

```python
x=x[:, 2]
print(x.shape)
```

Reshape to a 2-D array

```python
x=x.reshape((-1,1))
print(x.shape)
```

Split the data into training and testing data

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y, test_size=0.33)
```

Create a linear regression model and train it with our data

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression ()
model.fit(x_train, y_train)
```

Predict using our test data

```python
y_pred = model.predict(x_test)
```

```python
#create a scatter plot
plt.scatter(x_test, y_test, color = 'black')
#Plot the prediction
plt.plot(x_test, y_pred, color = 'blue', linewidth = 3)
#Add labels and a title
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Showing Diabetes Progression Against BMI')
#Show the plot
plt.show()
```
